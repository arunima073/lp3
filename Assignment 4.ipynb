Implement Gradient Descent Algorithm to find the local minima of a function.
For example, find the local minima of the function y=(x+3)Â² starting from the point x=2.

```python
#Initialize Parameters
cur_x = 2
rate = 0.01
precision = 0.000001
previous_step_size = 1
max_iters = 1000
iters = 0
df = lambda x : 2 * (x + 3) #Gradient of  function 
```

```python
#Run a loop to perform gradient Descent
while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x
    cur_x -= rate * df(prev_x)
    previous_step_size = abs(prev_x - cur_x)
    iters += 1
print("Local Minima Occurs at :",cur_x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return (x+3) ** 2
```

```python
def derivative(x):
    return 2 * (x+3)
```

```python
def gradient_descent(x_start, learning_rate, num_i):
    x_val=[]
    y_val=[]
    x_curr=x_start
    for _ in range(num_i):
        x_val.append(x_curr)
        y_val.append(func(x_curr))
        x_curr=x_curr - learning_rate * derivative(x_curr)
    return x_val, y_val
```

```python
x_start=2
learning_rate=0.1
num_i=50
# Run Gradient Descent
x_val, y_val = gradient_descent(x_start, learning_rate, num_i)
```

```python
x=np.linspace(-8,2,100)
y=func(x)
```

```python
plt.plot(x, y, label='y = (x + 3)^2')
plt.scatter(x_val, y_val, color='red', label='Gradient Descent Steps')
plt.title('Gradient Descent to Find Local Minima')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
