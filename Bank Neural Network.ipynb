```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
df = pd.read_csv("Churn_Modelling.csv")
```

```python
df.drop(["RowNumber","CustomerId","Surname"],axis=1,inplace=True)
```

```python
df.describe()
```

```python
states = pd.get_dummies(df["Geography"])
gender = pd.get_dummies(df["Gender"])
```

```python
df = pd.concat([df,states,gender], axis=1)
```

```python
df
```

```python
df.columns
```

```python
x = df[['CreditScore', 'Age', 'Tenure', 'Balance',
       'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary', 'France', 'Germany', 'Spain', 'Female', 'Male']]
```

```python
y = df["Exited"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=42)
```

```python
x
```

## Normalize Values

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
```

```python
x_train
```

```python
x_test = sc.transform(x_test)
```

```python
x_test
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(activation="relu", input_dim=13, units=6, kernel_initializer="uniform"))
```

```python
classifier.add(Dense(activation="relu", units=6, kernel_initializer="uniform"))
```

```python
classifier.add(Dense(activation="sigmoid", units=1, kernel_initializer="uniform"))
```

```python
classifier.compile(metrics=["accuracy"], optimizer="adam", loss="binary_crossentropy")
```

```python
classifier.summary()
```

```python
classifier.fit(x_train, y_train, batch_size=10, epochs=50)
```

```python
predict = classifier.predict(x_test)
```

```python
predict = (predict > 0.5)
```

```python
confusion_matrix(y_test, predict)
```

```python
accuracy_score(y_test, predict)
```

```python
print(classification_report(y_test, predict))
```
